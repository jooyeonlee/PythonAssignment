```python
import requests as r
from IPython.display import clear_output
# Instead of Making a Pokemon Dictionary, I want to make pokemon objects
# I want to store those pokemon objects in a dictionary where the key is the pokemon's name
# {
# 'grovyle' : <pokemon_object for grovyle @ 0x304180sflk31sj>
# }
# I want to be able to pass a dictionary made from the API call .json() data into the __init__() of Pokemon class
# and have the pokemon's attributes be filled out from there

# let me lay out my skeleton code
# pokemon object is gonna have the same attributes
    # name=str, abilities=[], types=[], weight=int
# pokemon object methods
    # display that prints our pokemon's info nice and pretty prettily? fancy-lookin.

# second class pokedex
    # 1 attribute - the dictionary of all the pokemon
    
    # 3 methods
        # 1 create pokemon -> take in a list of pokemon names, and fill up our objects/dictionary
        # 2 display function - to show all the pokemon
        # 3 searching function to display based on the pokemon type asked for
        
         
# pokemon objects
class Pokemon:
    def __init__(self, data):
        self.name = data['name']
        self.abilities = [ab['ability']['name'] for ab in data['abilities']]
        self.types = [ab['type']['name'] for ab in data['types']]
        self.weight = data['weight']
    def printInfo(self):
        abilities = (",").join(self.abilities)
        types = (",").join(self.types)
        print(f"{self.name.title()} weighs {self.weight}lb and is a type {types} Pokemon with {abilities}")

# pokedex objects
class Pokedex:
    def __init__(self):
        self.dict = {}
    def createPokemon(self, pokename):
        data = r.get(f"https://pokeapi.co/api/v2/pokemon/{pokename}")
        if data.status_code == 200:
            my_data = data.json()
            pokemon = Pokemon(my_data)
            self.dict[pokename] = pokemon
        else:
            print("Fail to create")
    def showAll(self):
        for value in self.dict.values():
            value.printInfo()
    def searchType(self, poketype):
        typelist = []
        for key, value in self.dict.items():
            for tp in value.types:
                if tp == poketype:
                    typelist.append(value)
        if len(typelist) > 0:
            for tp in typelist:
                tp.printInfo()
        else:
            print(f"There isn't Pokemon of type {poketype}")
    def createPokeList(self):
        data = r.get("https://pokeapi.co/api/v2/pokemon/")
        if data.status_code == 200:
            my_data = data.json()
            pokelist = [mon['name'] for mon in my_data['results']]
            return pokelist
        else:
            print("Fail to create pokemon list")
            return None
            
        
class Main:
    @classmethod 
    def run(self):
        pokedex = Pokedex()
        pokelist = pokedex.createPokeList()
        if len(pokelist) != 0:
            for poke in pokelist:
                pokedex.createPokemon(poke)
                
        while True:
            answer = input("What type of pokemon do you want to search for? type/all for show all/quit")
            clear_output()

            if answer.lower() == "all":
                pokedex.showAll()
            elif answer.lower() == "quit":
                break
            else:
                pokedex.searchType(answer.lower())

Main.run()
```
