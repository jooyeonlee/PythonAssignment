```python
from IPython.display import clear_output

class Property:
    def __init__(self):
        self.income = 0
        self.tax = 0
        self.insurance = 0
        self.vacancy = 0
        self.morgage = 0
        self.repair = 0
        self.capex = 0
        self.cashflow = 0
        self.downpayment = 0
        self.closingcost = 0
        self.rehabbudget = 0
        self.totalexpense = 0
        self.cashflow = 0
        self.totalinvest = 0
        self.roi = 0
    
    def inputincome(self):
        print("=============Income==============")
        try:
            income = int(input("How much monthly income do you generate with this property?"))
            self.income = income
            print("------------------------------")
            print(f"Your total income is {self.income}")
        except ValueError:
            print("Please enter only numbers")
            
    def inputexpense(self):
        print("=============Expense==============")
        try:
            tax = int(input("How much is the monthly tax for the property?"))
            self.tax = tax
            insurance = int(input("How much is monthly insurance for the property?"))
            self.insurance = insurance
            vacancy = int(input("What is the amount due to vacancy each month?"))
            self.vacancy = vacancy
            morgage = int(input("How much do you pay for the morgage each month?")) 
            self.morgage = morgage
            repair = int(input("How much does a monthly repair cost?"))
            self.repair = repair
            capex = int(input("How much is your monthly capex?"))
            self.capex = capex
            self.totalexpense = self.tax + self.insurance + self.vacancy + self.morgage + self.repair + self.capex
            print("------------------------------")
            print(f"Your total total expense is {self.totalexpense}")
        except ValueError:
            print("Please enter only numbers")
        
    def inputinvest(self):
        print("=============Investment==============")
        try:
            downpayment = int(input("How much down payment do you pay?"))
            self.downpayment = downpayment
            closingcost = int(input("How much is the closing cost?"))
            self.closingcost = closingcost
            rehabbudget = int(input("What is your budget for rehabilitation?"))
            self.rehabbudget = rehabbudget
            self.totalinvest = self.downpayment + self.closingcost + self.rehabbudget
            print("------------------------------")
            print(f"Your total total investment is {self.totalinvest}")
        except ValueError:
            print("Please enter only numbers")
    
    def calcashflow(self):
        self.cashflow = self.income - self.totalexpense
        return self.cashflow
        
    def calROI(self):
        try:
            self.roi = (self.cashflow/self.totalinvest) * 100
            return self.roi
        except ZeroDivisionError:
            return None
    
    def showproperty(self, name):
        print(f"{name} ROI is {self.roi:.2f}%")
        print("ROI is calculated based on the information:")
        print(f"Income: {self.income} | Tax: {self.tax} | Insurance: {self.insurance} | Vacancy: {self.vacancy} | Repair: {self.repair} | Capex: {self.capex} | Down Payment: {self.downpayment} | Closing Cost: { self.closingcost} | Budget: {self.rehabbudget}")

class PropertyList:
    def __init__(self):
        self.properdict = {}
    
    def addProperty(self, name, realestate):       
        self.properdict[name] = realestate
        
    def searchProperty(self, name):
        for key, value in self.properdict.items():
            if key == name:
                value.showproperty(name)
                return
        print("There is no record of calculating with the property.")
                
    def showallrecord(self):
        record = [key for key in self.properdict.keys()]
        return record
        
    def findrecord(self, name):
        for key in self.properdict.keys():
            if key == name:
                return True
        return False
    
    def createproperty(self, prop):
        prop.inputincome()
        prop.inputexpense()
        prop.inputinvest()
        cashflow = prop.calcashflow()
        roi = prop.calROI()
        return roi
    
class Main:
    @classmethod 
    def run(self):
        print("Welcome to Rental ROI Calculator")
        record = PropertyList()
        
        while True:
            want = input("What do you want to do? Input 's' for search / 'c' for new calculation / 'q' for quit")
            clear_output()
            
            if want == 's':
                result = record.showallrecord()
                if len(result):
                    print("There are records of calculations for the following properties:" + ",".join(result))
                    name = input("Please enter the property name that you want to search:")
                    record.searchProperty(name)
                else:
                    print("There isn't any record")

            elif want == 'c':
                name = input("Please enter the property name:")
            
                if record.findrecord(name):
                    print("There is a record of calculations for that house.")
                    modify = input("Do you want to modify the information? 'y' for yes 'n' for no")
                    if modify == 'n':
                        continue
                
                prop = Property()
                roi = record.createproperty(prop)
                if roi == None:
                    print("You cannot calculate ROI with invalid value")
                else:
                    print(f"{name} ROI is {roi:.2f}%")
                    record.addProperty(name, prop)
            elif want == 'q':
                print("See you next time")
                break
            else:
                print(f"{answer} is not valid input.")
                
Main.run()
```
