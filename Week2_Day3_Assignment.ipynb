### 1) Build a Shopping Cart <br>
<p><b>You can use either lists or dictionaries. The program should have the following capabilities:</b><br><br>
1) Takes in input <br>
2) Stores user input into a dictionary or list <br>
3) The User can add or delete items <br>
4) The User can see current shopping list <br>
5) The program Loops until user 'quits' <br>
6) Upon quiting the program, print out all items in the user's list <br>
</p>

```python
from IPython.display import clear_output

# Shopping cart dictionary {(String)item: (int)item number}
cart = {}

# Add item
def additem():
    # Input item name and number of the item to add
    item = input("What do you want to add into you shopping list?").title()
    try:
        itemnum = int(input("How many items do you want to add?"))
        # if the item is in cart already, number of item add to the originally stored number
        if item in cart.keys():
            temp = cart[item]
            cart[item] = temp + itemnum
        else:
        # if the item is not in the cart, add new key and value
            cart[item] = itemnum
        clear_output()
        print(f"{itemnum} {item} has been added to your cart")
    except ValueError:
        print("Please a number of item. You cannot enter string")
        

# Delete item
def delitem():
    # Input item name and number of the item to delete
    item = input("What you want to delete from your shopping list").title()
    try:
        itemnum = int(input("How many items do you want to remove?"))
        clear_output()
        # if the item is in cart already, number of item deduct from the originally stored number
        if item in cart.keys():
            if itemnum < cart[item]:
                temp = cart[item]
                cart[item] = temp - itemnum
                print(f"{itemnum} {item} has been removed from your shopping cart")
            else:
                #if the number of item to delete is larger than the orinigal number, delete the key and value from dictionary
                del cart[item]
                print(f"All of {item} has been removed from your shopping cart")
        else:
            print(f"{item} is not your cart, so it cannot be deleted ")
    except ValueError:
        print("Please a number of item. You cannot enter string")    

# Show item when show or quit
def showitem():
    if len(cart) == 0:
        print("Your shopping cart is empty")
    else:
        print(f"Your shopping cart list: {cart}")

def shoppingcart():
    
    while True:
        answer = input("Enter what you want to do (Show/Add/Delete/Clear or Quit)")
        clear_output()
        
        if answer.lower() == "show":
            showitem()
        elif answer.lower() == "add":
            additem()
        elif answer.lower() == "delete":
            delitem()
        elif answer.lower() == "clear":
            cart.clear()
            print("Your all of shopping cart items has been removed")
        elif answer.lower() == "quit":
            showitem()
            break
        else:
            print(f"{answer} is not valid input.")


shoppingcart()
```
