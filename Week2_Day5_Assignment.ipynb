### Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program

The comments in the cell below are there as a guide for thinking about the problem. However, if you feel a different way is best for you and your own thought process, please do what feels best for you by all means.

```python
# Create a class called cart that retains items and has methods to add, remove, and show
from IPython.display import clear_output

# Item class
class Item():
    def __init__(self, name, price, amount):
        self.name = name
        self.price = price
        self.amount = amount
        
    def displayInfo(self):
        print(f"{self.name}: ${self.price} / {self.amount}")

# Cart class for storing shopping list
class Cart():
    def __init__(self):
        # Shopping cart list
        self.cart = {}
        self.total = 0
            
    # Add Item
    def addItem(self, item, price):
        #if the item in the cart, the number of item increase by 1
        if item in self.cart.keys():
            temp = self.cart[item]
            self.cart[item] = temp + 1
        # if the item is not in the cart, add new key and value
        else:
            self.cart[item] = 1
        self.total = self.total + price
        print(f"{item} has been added to your cart")
        
    # Delete item
    def deleteItem(self, item, price):
        # if the item is in cart already, the number of the item decrease by 1
        if item in self.cart.keys():
            if self.cart[item] > 1:
                temp = self.cart[item]
                self.cart[item] = temp - 1
                print(f"1 {item} has been removed from your cart: now you have {self.cart[item]} {item}s")
            else:
                #if the number of item is 1, delete the key and value from dictionary
                del self.cart[item]
                print(f"{item} has been removed from your shopping cart")
            self.total = self.total - price
        # if the item isn't in the cart
        else:
            print(f"{item} is not your cart, so it cannot be deleted ")   
    
    # Show cart items
    def showItem(self):
        if len(self.cart) == 0:
            print("Your shopping cart is empty")
        else:
            string = ""
            for key, value in self.cart.items():
                string += f"{key}: {value} "
            print(f"Your shopping cart list: {string}")
            print(f"Your total cost is ${self.total:.2f}")
    
    # Clear Cart: Remove all of items from the cart dictionary
    def clearItem(self):
        self.cart.clear()
        self.total = 0
        print("Your all of shopping cart items has been removed")

# store class to manage stocks of items
class Store():
    def __init__(self):
        #class Item list
        self.stocks = []
    
    # add item to stock list
    def addStock(self, item):
        self.stocks.append(item)

    # check if the item in the stock list
    def checkStock(self, item):
        for stock in self.stocks:
            if item == stock.name:
                return True
        return False
    
    # show the stock list 
    def showStock(self):
        item_list = [stock.name for stock in self.stocks]
        itemstr = ', '
        return itemstr.join(item_list)
    
    # show the information about the item in stock
    def getStockInfo(self, item):
        for stock in self.stocks:
            if item == stock.name:
                stock.displayInfo()
                return
        print(f"Sorry, we don't have {item} now. Please enter the items currently in the store")
        
    def getPrice(self, item):
        for stock in self.stocks:
            if item == stock.name:
                return stock.price
        return 0
    
class Main:
    @classmethod 
    def run(self):    
        store = Store()
        apple = Item('Apple', 3.69, '1lb')
        milk = Item('Milk', 3.59,'64fl oz')
        egg = Item('Egg', 4.99, '12 count')
        avocado = Item('Avocado', 2.49, '1lb')
        coffee = Item('Coffee', 9.89, '12oz')
        store.addStock(apple)
        store.addStock(milk)
        store.addStock(egg)
        store.addStock(avocado)
        store.addStock(coffee)
        
        cart = Cart()
        
        print(f"Hello, welcome to the Store. We have {store.showStock()}")
    
        while True:
            
            answer = input("Enter what you want to do (Show/Add/Delete/Clear/Info or Quit)")
            clear_output()

            if answer.lower() == "show":
                cart.showItem()
            elif answer.lower() == "add":
                # Input item name to add
                print(f"We have {store.showStock()}")
                item = input("What do you want to add into you shopping list?").title()
                if store.checkStock(item):
                    price = store.getPrice(item)
                    cart.addItem(item, price)
                else:
                    print(f"Sorry, we don't have {item} now. Please enter the items currently in the store")
            elif answer.lower() == "delete":
                item = input("What you want to delete from your shopping list").title()
                price = store.getPrice(item)
                cart.deleteItem(item, price)
            elif answer.lower() == "clear":
                cart.clearItem()
            elif answer.lower() == "info":
                item = input("What item do you want information about?").title()
                store.getStockInfo(item)
            elif answer.lower() == "quit":
                cart.showItem()
                break
            else:
                print(f"{answer} is not valid input.")

Main.run()
```
